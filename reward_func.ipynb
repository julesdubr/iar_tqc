```python
import gym
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from single_mdp import MDP
```

```python
n = 50

env = MDP()
actions = np.linspace(-1, 1, n)
rewards = np.zeros(n)
targets = np.zeros(n)
```

```python
env.reset()

for i, action in enumerate(actions):
    a = action.item()
    _, reward , _, _ = env.step(a)
    targets[i] = env.mean_reward(a)
    rewards[i] = reward
```

```python
custom_params = {
    "axes.grid": True,
    "axes.spines.right": False,
    "axes.spines.top": False,
}
sns.set_theme(style="ticks", palette="tab10", rc=custom_params)

plt.plot(actions, targets, color="r", linewidth=3, label='$f(a)$')
plt.scatter(actions, rewards, label='samples from $R(a)$')

plt.ylim(ymin=-2)
plt.legend()
plt.show()
```
